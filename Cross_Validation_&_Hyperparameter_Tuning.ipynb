# <b>A. <span style='color:#3A6D8C'>Introduction</span></b>

Analisis regresi adalah teknik statistik yang digunakan untuk memahami hubungan antara satu atau lebih variabel independen dan satu variabel dependen. Tujuannya adalah untuk memodelkan dan memprediksi nilai variabel dependen berdasarkan nilai variabel independen. Ada beberapa jenis analisis regresi, seperti:

* Regresi Linear Sederhana: Menganalisis hubungan antara satu variabel independen dan satu variabel dependen dengan menggunakan garis lurus.
* Regresi Linear Berganda: Menganalisis hubungan antara beberapa variabel independen dan satu variabel dependen.
* Regresi Non-Linear: Digunakan ketika hubungan antara variabel tidak linier.
* Regresi Logistik: Digunakan ketika variabel dependen bersifat kategorikal, seperti ya/tidak.

Dengan analisis regresi, kita bisa mengidentifikasi faktor-faktor yang mempengaruhi hasil tertentu dan membuat prediksi berdasarkan data yang ada.

### <b>C.3. <span style='color:#55679C'>Variance Inflation Factor (VIF)</span></b>

Variance Inflation Factor (VIF) adalah ukuran yang digunakan untuk mendeteksi multikolinearitas dalam model regresi. Multikolinearitas terjadi ketika dua atau lebih variabel independen dalam model regresi memiliki korelasi yang tinggi satu sama lain, yang dapat menyebabkan masalah dalam estimasi koefisien regresi.
<br><br>
<table>
    <thead>
        <tr>
            <th>Nilai VIF</th>
            <th>Interpretasi</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>VIF = 1</td>
            <td>Tidak ada korelasi antara variabel independen tersebut dan yang lainnya.</td>
        </tr>
        <tr>
            <td>1 < VIF < 5</td>
            <td>Korelasi rendah, tidak ada masalah serius.</td>
        </tr>
        <tr>
            <td>5 <= VIF < 10</td>
            <td>Menandakan potensi masalah multikolinearitas.</td>
        </tr>
        <tr>
            <td>VIF  10</td>
            <td>Menunjukkan multikolinearitas yang serius dan biasanya dianggap sebagai batasan yang harus diperhatikan.</td>
        </tr>
    </tbody>
</table>
<br><br>

VIF Digunakan untuk
* `Stabilitas Model`: Multikolinearitas dapat membuat estimasi koefisien regresi tidak stabil dan sulit untuk diinterpretasikan.
* `Pengujian Hipotesis`: Mempengaruhi hasil pengujian hipotesis, yang dapat menyesatkan dalam pengambilan keputusan.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools import add_constant

# Menambahkan konstanta untuk VIF
X = add_constant(number_data)

# Menghitung VIF untuk setiap variabel
vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_data = vif_data.loc[vif_data['Variable'] != 'const']

# Tipe VIF
condition_VIF = [
    vif_data['VIF'] == 1,
    (1 < vif_data['VIF']) & (vif_data['VIF'] < 5),
    (5 <= vif_data['VIF']) & (vif_data['VIF'] < 10),
]

values_VIF = [
    'Tidak Ada Korelasi',
    'Korelasi Rendah',
    'Berpotensi Multikolinearitas',
]

vif_data['tipe VIF'] = np.select(condition_VIF, values_VIF, default = 'Multikolinearitas Serius')

# Tampilkan hasilnya
display(vif_data)
```

```python
def replace_outliers_iqr(df):

    # Menghitung Q1 dan Q3
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1

    # Menentukan batas bawah dan atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Mengganti outlier
    df = np.select(
        [df > upper_bound, df < lower_bound],
        [upper_bound, lower_bound],
        default = df
    )

    return df
```

```python
for col in number_data.columns :
    number_data[col] = replace_outliers_iqr(number_data[col])
```

### <b>D.1. <span style='color:#55679C'>Definisikan Variabel Bebas & Variabel Terikat</span></b>

Variabel adalah suatu atribut, sifat atau nilai dari orang, serta objek atau kegiatan yang memiliki variasi tertentu untuk dipelajari dan ditarik kesimpulannya. Secara umum variabel dibagi menjadi 2 :

* `Variabel Bebas` (Independent Variable) : Variabel yang mempengaruhi, atau yang menjadi sebab perubahan dari adanya suatu variabel dependen (terikat). Variabel bebas biasanya dinotasikan dengan X.

* `Variabel Terikat` (Independent Variable) : Variabel yang dipengaruhi, akibat adanya variabel bebas. Variabel ini biasa dinotasikan dengan Y.

Untuk data_kemiskinan yang jadi variabel terikat adalah `presentase_penduduk_miskin` (variabel yang akan diteliti) dan yang jadi variabel bebas adalah *angka_harapan_hidup, pengeluaran_per_kapita, rataan_lama_sekolah dan presentase_penduduk_merokok* untuk diteliti pengaruhnya terhada tingkat kemiskinan di Indonesia


```python
# Definisikan variabel bebas dan terikat
X = number_data.drop(columns = 'presentase_penduduk_miskin')
y = number_data[['presentase_penduduk_miskin']]
```

### <b>D.2. <span style='color:#55679C'>Lakukan Proses Scaling</span></b>

Setelah ditentukan variabel bebas dan terikatnya langkah selanjutnya adalah proses `Scaling`. Proses ini bertujuan untuk mengubah rentang atau distribusi data sehingga nilai-nilainya berada dalam skala tertentu. Perhatikan data berikut :

```python
# Tampilkan data
number_data.head()
```

Terlihat bahwa baik variabel terikat maupun variabel bebas rentang nilainya tidak pada rentang yang sama, sehingga perlu untuk `disetarakan` terlebih dahulu dengan proses *scaling*. Pada kasus ini akan digunakan metode Min-Max Scaler. `Min-Max Scaler` mengubah semua fitur sehingga berada dalam rentang 0 dan 1. Metode ini memastikan bahwa semua fitur berkontribusi secara seimbang dalam model, menghindari dominasi fitur dengan rentang nilai yang lebih besar.

```python
# Import library yang dibutuhkan
from sklearn.preprocessing import MinMaxScaler

# Inisialisasi scaler
min_max_scaler = MinMaxScaler()

# Melakukan scaling
X_scaled = min_max_scaler.fit_transform(X)

# Mengembalikan hasil scaling ke dalam DataFrame
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)

# Tampilkan hasilnya
display(X_scaled.head())
```

```python
# Melakukan scaling
y_scaled = min_max_scaler.fit_transform(y)

# Mengembalikan hasil scaling ke dalam DataFrame
y_scaled = pd.DataFrame(y_scaled, columns = y.columns)

# Tampilkan hasilnya
display(y_scaled.head())
```

### <b>D.3. <span style='color:#55679C'>Pendefinisian Model</span></b>

Model yang digunakan untuk regresi adalah XGBoost. XGBoost (diklaim) sangat baik dalam menangani pola data non-linear karena menggunakan pendekatan berbasis pohon keputusan. Kemampuan pohon keputusan dalam menangkap hubungan non-linear, digabungkan dengan metode boosting dan regularisasi, menjadikannya pilihan yang kuat untuk berbagai masalah prediksi yang kompleks.

```python
# Import library yang digunakan
from xgboost import XGBRegressor

# Buat objek XGBoost
modelXGB = XGBRegressor(objective='reg:squarederror')

# Tampilkan hasil objek
display(modelXGB)
```

### <b>D.4. <span style='color:#55679C'>Pendefinisian Metrik Evaluasi</span></b>

Untuk mengukur kinerja model regresi salah satunya dengan menggunakan Mean Absolute Error (MAE) sebagai metrik evaluasi. MAE dirumuskan sebagai :

\begin{equation}
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
\end{equation}



```python
# Import library yang digunakan
from sklearn.metrics import make_scorer, mean_absolute_error

# Definisikan MAE
neg_mae_scorer = make_scorer(
    mean_absolute_error,
    greater_is_better = False
)
```

`mean_absolute_error` : Fungsi untuk menghitung rata-rata selisih absolut antara nilai prediksi dan nilai sebenarnya. Semakin rendah nilainya, semakin baik kinerja model.

`greater_is_better = False` : Skor yang lebih rendah (yang menunjukkan MAE lebih rendah) dianggap lebih baik.

### <b>D.5. <span style='color:#55679C'>Cross Validation (CV)</span></b>

Cross-validation adalah teknik evaluasi model yang digunakan untuk menilai seberapa baik model machine learning dapat digeneralisasi ke data yang tidak terlihat. Teknik ini membantu menghindari overfitting, yaitu ketika model terlalu kompleks dan hanya cocok dengan data pelatihan, tetapi berkinerja buruk pada data baru.
<br><br>
<img src="https://miro.medium.com/v2/resize:fit:1200/1*AAwIlHM8TpAVe4l2FihNUQ.png" width="60%">

<br>

<b>Manfaat Melakukan CV</b>

* `Evaluasi yang Lebih Akurat`: Memberikan estimasi yang lebih baik tentang seberapa baik model akan berkinerja pada data yang tidak terlihat.
* `Mengurangi Overfitting`: Dengan menggunakan data yang berbeda untuk pelatihan dan pengujian, cross-validation membantu meminimalkan risiko model hanya cocok dengan data pelatihan.
* `Pemilihan Model`: Membantu dalam memilih model yang paling sesuai untuk dataset tertentu berdasarkan kinerja rata-rata selama validasi.

```python
# Import library
from sklearn.model_selection import cross_val_score

# Inisialisasi jumlah CV
jumlah_cv = 10

# Perform k-fold cross-validation
scores = cross_val_score(
    modelXGB,
    X_scaled, y_scaled,
    cv = jumlah_cv,
    scoring = neg_mae_scorer
)

# Ubah ke bentuk positif
scores = -scores
```

```python
import plotly.express as px

fig = px.bar(
    x = ['Percobaan ' + str(i+1) for i in range(len(scores))],
    y = scores,
    color_discrete_sequence = ['#FA8282']
)

mean_value = pd.Series(scores).mean()
std_value = pd.Series(scores).std()

fig.add_hline(
    y = mean_value,
    line_dash = 'dash',
    line_color = 'red',
    annotation_text = f'X = {mean_value:.2f}<br> = {std_value:.2f}',
    annotation_position = "bottom right"
)

# Percantik tampilan
fig.update_layout(
    width = 1200,
    height = 600,
    plot_bgcolor = 'rgba(0, 0, 0, 0)',
    xaxis_title = 'Percobaan',
    yaxis_title = 'Error',
    title = dict(
        text = f"<b>Hasil Cross Validation dengan {jumlah_cv} Percobaan</b><br><sup>Pada Model XGBoost</sup>",
        font = dict(
            size = 25,
            color = '#0E2954'
        ),
        y = 0.94,
        x = 0.5
    )
)

fig.show(renderer = 'colab')
```

<b><span style='color:#FFDB00'><i>Interpretasi Hasil : </i></span></b><br>
Dari hasil 10 kali percobaan Cross Validation pada model XGBoost, diperoleh rata-rata MAE (Mean Absolute Error) sebesar 0.16 dengan simpangan baku 0.09

Rata-rata MAE dan simpangan baku bersama-sama memberikan gambaran tentang seberapa baik model bekerja. Dalam kasus ini Rata-rata MAE yang cukup rendah (0.16) menunjukkan model melakukan prediksi yang baik. Simpangan baku (0.09) yang relatif kecil menunjukkan bahwa performa model konsisten di berbagai percobaan.

<ul>
  <li><b>Underfit</b></li>

Akibatnya, model bekerja dengan sangat baik pada data pelatihan tetapi gagal untuk menggeneralisasi dengan baik pada data baru atau data yang belum terlihat (data validasi atau test).<br><br>

```python
import numpy as np

# Buat proporsi data latih
proporsi_train = np.linspace(0.1, 1.0, 10)

# Tampilkan hasilnya
display(proporsi_train)
```

```python
from sklearn.model_selection import learning_curve

# Menghitung learning curve
train_sizes, train_scores, test_scores = learning_curve(
    modelXGB,
    X_scaled, y_scaled,
    cv = jumlah_cv,
    scoring = neg_mae_scorer,
    train_sizes = proporsi_train
)
```

```python
pd.set_option('display.float_format', '{:.6f}'.format)

# Rata - rata error train & test
train_scores_mean = np.abs(np.mean(train_scores, axis=1))
test_scores_mean = np.abs(np.mean(test_scores, axis=1))

# Buat dataframenya
pd.DataFrame({
    'train_sizes' : train_sizes,
    'test_sizes' : X.shape[0] - train_sizes,
    'train_scores_mean' : train_scores_mean,
    'test_scores_mean' : test_scores_mean
})
```

```python
import pandas as pd
import plotly.express as px

# Membuat DataFrame
df = pd.DataFrame({
    'train_sizes': train_sizes,
    'train_scores_mean': train_scores_mean,
    'test_scores_mean': test_scores_mean
})

# Membuat plot line
fig = px.line(df,
              x='train_sizes',
              y=['train_scores_mean', 'test_scores_mean'],
              labels={'value': 'Score', 'train_sizes': 'Training Sizes'},
              title='Learning Curve: Training and Validation Scores')

# Menambahkan keterangan
fig.update_traces(mode='lines+markers')  # Menampilkan garis dan marker

# Percantik tampilan
fig.update_layout(
    width = 1200,
    height = 600,
    plot_bgcolor = 'rgba(0, 0, 0, 0)',
    title = dict(
        text = f"<b>Learning Curves</b><br><sup><sup>Pada Model XGBoost</sup></sup>",
        font = dict(
            size = 25,
            color = '#0E2954'
        ),
        y = 0.94,
        x = 0.5
    )
)

# Menampilkan plot
fig.show(renderer = 'colab')
```

Terdapat kesenjangan yang cukup besar antara skor pelatihan dan pengujian pada ukuran set pelatihan yang kecil, menunjukkan adanya `overfitting`, walaupun jarak antara kedua skor ini semakin mengecil seiring bertambahnya ukuran set pelatihan. Sehingga perlu untuk dilakukan tuning parameter agar model dapat mengeneralisir data yang dimodelkan.

### <b>D.7. <span style='color:#55679C'>Hyperparameter Tuning</span></b>

Hyperparameter tuning adalah proses mengoptimalkan hyperparameter model machine learning untuk meningkatkan kinerja dan akurasi model. Hyperparameter adalah parameter yang tidak dioptimalkan secara langsung melalui pembelajaran, melainkan ditentukan sebelum pelatihan model.

<img src="https://raw.githubusercontent.com/bachtiyarma/Material/refs/heads/main/Image/Materi-Python/XGBoost%20Documentation.png" width="70%">

docs : <i>https://xgboost.readthedocs.io/en/stable/python/python_api.html#xgboost.XGBRegressor</i>

```python
from sklearn.model_selection import learning_curve

jumlah_cv = 5

for alpha in [1, 2, 3]:
    for lambdas in [1, 2, 3]:
        for learn_rate in [0, 0.2, 0.4, 0.6]:
            # Buat objek XGBoost
            modelXGB_tuned = XGBRegressor(
                objective = 'reg:squarederror',
                max_depth = 3,
                max_leaves = 2,
                reg_alpha = alpha,
                reg_lambda = lambdas,
                learning_rate = learn_rate
            )

            # Menghitung learning curve
            train_sizes_tuned, train_scores_tuned, test_scores_tuned = learning_curve(
                modelXGB_tuned,
                X_scaled, y_scaled,
                cv = jumlah_cv,
                scoring = neg_mae_scorer,
                train_sizes = proporsi_train
            )

            # Rata - rata error train & test
            train_scores_mean_tuned = np.abs(np.mean(train_scores_tuned, axis=1))
            test_scores_mean_tuned = np.abs(np.mean(test_scores_tuned, axis=1))

            # Membuat DataFrame
            df_tuned = pd.DataFrame({
                'train_sizes_tuned': train_sizes_tuned,
                'train_scores_mean_tuned': train_scores_mean_tuned,
                'test_scores_mean_tuned': test_scores_mean_tuned
            })

            # Membuat plot line
            fig = px.line(
                df_tuned,
                x = 'train_sizes_tuned',
                y = ['train_scores_mean_tuned', 'test_scores_mean_tuned'],
                labels = {'value': 'Score', 'train_sizes_tuned': 'Training Sizes'},
            )

            # Menambahkan keterangan
            fig.update_traces(
                mode = 'lines+markers'
            )

            # Percantik tampilan
            fig.update_layout(
                width = 1000,
                height = 400,
                plot_bgcolor = 'rgba(0, 0, 0, 0)',
                title = dict(
                    text = f"<b>Learning Curves Pada Model XGBoost</b><br><sup><sup>Parameter (alpha : {alpha}, lambda : {lambdas}, learning_rate : {learn_rate})</sup></sup>",
                    font = dict(
                        size = 25,
                        color = '#0E2954'
                    ),
                    y = 0.94,
                    x = 0.5
                )
            )

            # Menampilkan plot
            fig.show(renderer = 'colab')
```

Dari hasil percobaan diatas diperoleh bahwa parameter terbaik adalah reg_alpha = 1, reg_lambda = 3 dan learning_rate = 0.6. Hal ini dikarenakan error pada data test menunjukan trend penurunan begitu juga dengan data train.
